- Here is how I rate my first experience working in ArcGIS Pro
  - Data important and management
    - I thought that this was one of the more simpler parts of creating the map. Being able to manage all of the important data and transfer it from one layout to another was relatively easy.
  - Map creation and annotation
    - Similar to the data important and management, I thought that this was relatively simple and easy to navigate. It was easy to insert all items and be able to create a map that has life to it.
