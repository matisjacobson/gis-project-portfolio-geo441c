## **Reflection on data management function in ArcGIS Pro**

- **Write a 100-word paragraph to describe the relationship between**
  - **`Geodatabase`**
  - **`Feature Dataset`**
  - **`Feature Classes`**
  - I can refer to **Page 9** of [assignment guide](https://miamioh.instructure.com/courses/246872/files/folder/week-02-resources?preview=38195670) or other resources online, such as
    - [Fundamentals of geodatabase](https://pro.arcgis.com/en/pro-app/latest/help/data/geodatabases/overview/fundamentals-of-the-geodatabase.htm)
    - [Feature datasets in ArcGIS Pro](https://pro.arcgis.com/en/pro-app/latest/help/data/feature-datasets/feature-datasets-in-arcgis-pro.htm)
- A geodatabase is the main "container" that is used in ArcGIS Pro to store and manage data, taables, and all of the relationships between datasets.
- Inside of the geodatabase, a feature dataset acts as a folder that groups related feature classes that share the same coordinate system and are often used together.
- Within the feature dataset, feature classes store the actual geographic features, such as the points, lines, or polygons, along with their attribute data.
- This structure helps to organize spacial information more efficiently and ensures data consistency. It also supports topology and relationships, and improves data management when working with multiple related geographic layers in ArcGIS Pro.
